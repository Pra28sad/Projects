```python
# import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics

import warnings
warnings.filterwarnings('ignore')
```

# Data Preprocessing

```python
# Extract brand and model from CarName
cars['brand'] = cars['CarName'].apply(lambda x: x.split(' ')[0])
cars['model'] = cars['CarName'].apply(lambda x: ' '.join(x.split(' ')[1:]))

# Define categorical and numerical columns
label_encoding_columns = ['carbody', 'enginetype', 'cylindernumber', 'fuelsystem', 'brand', 'model']
one_hot_encoding_columns = ['fueltype', 'aspiration', 'drivewheel', 'enginelocation']
numerical_columns = ['wheelbase', 'carlength', 'carwidth', 'carheight', 'curbweight',
                     'enginesize', 'boreratio', 'stroke', 'compressionratio', 'horsepower',
                     'peakrpm', 'citympg', 'highwaympg']

# Converting string numbers to actual numbers
cars['cylindernumber'] = cars['cylindernumber'].map({'four': 4, 'six': 6, 'five': 5, 'three': 3, 'twelve': 12, 'two': 2, 'eight': 8})
cars['doornumber'] = cars['doornumber'].map({'four': 4, 'two': 2})

# Encoding categorical variables using OneHotEncoding
cars = pd.get_dummies(cars, columns=one_hot_encoding_columns, drop_first=True, dtype=int)

# Encoding categorical variables using LabelEncoding
label_encoder = LabelEncoder()
for column in label_encoding_columns:
    cars[column] = label_encoder.fit_transform(cars[column])

# Encoding categorical variables using OneHotEncoding
# cars = pd.get_dummies(cars.drop(["CarName"], axis=1), drop_first=True)

# Feature scaling
scaler = StandardScaler()
cars[numerical_columns] = scaler.fit_transform(cars[numerical_columns])
```

# Train the model

```python
# Train test split
X = cars.drop(["price", "CarName"], axis=1)
y = cars["price"]

X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.7, random_state=42)

# Train the model
model = LinearRegression()
model.fit(X_train, y_train)

# Predictions
y_pred = model.predict(X_test)
```

```python
# Prediction summary
coefficients = pd.DataFrame(data = model.coef_, index = X.columns)
print(f"Coefficients are {coefficients}")
print(f"Intercept is {model.intercept_}")
print(f"R2 square value is {metrics.r2_score(y_pred, y_test)}")
print(f"Mean squared error is {metrics.mean_squared_error(y_pred, y_test)}")
print(f"Root mean squared error is {np.sqrt(metrics.mean_squared_error(y_pred, y_test))}")
print(f"Mean absolute error is {metrics.mean_absolute_error(y_pred, y_test)}")
```

```python
# Visualize the prediction vs actual results
plt.figure(figsize=(6, 4))
sns.scatterplot(x=y_test, y=y_pred)
plt.xlabel("Actual values")
plt.ylabel("Predicted values")
plt.title("Actual Vs Predicted")
plt.grid(True)
plt.show()
```
